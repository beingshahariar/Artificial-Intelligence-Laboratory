```python
# Part A: List and Tuple Operations
#A.1:Create a list of 10 random integers between 1 and 100
import random
numbers = [random.randint(1, 100) for _ in range(10)]
print('original list: ', numbers)
```

```python
#A.2:
#Sort the list in ascending and descending order
asc = sorted(numbers)
desc = sorted(numbers, reverse=True)
print('Asc order: ',asc)
print('Desc order: ',desc)
```

```python
# Remove duplicate values (if any).
unique = list(set(numbers))
print ('After removing duplicate: ',unique)
# Find and print the maximum and minimum values.
print('Maximum value: ',max(numbers))
print('Minimum value: ',min(numbers))
# Calculate and print the average.
avg = sum(numbers)/len(numbers)
print('Average: ',avg)
```

```python
# Part B: Set Operations
# B.1. Create two sets
setA = set(range(1,11))
setB = set(x for x in range(5,16)if x%2==0)
print('Set A: ',setA)
print('Set B: ',setB)
```

```python
# B.2. Perform and print the result of
# Union
# Intersection
# Difference (A- B and B- A)
# Symmetric difference
union_set = setA | setB
print('Union: ',union_set)

intersection = setA & setB
print('Intersection: ',intersection)

diff_A_B = setA - setB
print('Difference: ',diff_A_B)

diff_B_A = setB - setA
print('Difference: ',diff_B_A)

sym = setA ^ setB
print('Symmetric: ',sym)
```

```python
 # Part C: Dictionary Manipulation
 #C.1. Create a dictionary representing three students and their scores
students ={
   "Alice":85,
   "Bob":92,
   "Charlie":78  
}
print ('Original Dictonary: ',students)
print (students.values())
print(students.keys())
```

```python
# Add a new student "David" with score 88.
students['David'] = 88
print('After Adding David: ',students)

#Update "Alice"’s score to 90.
students['Alice'] = 90
print('After updateing Alice: ',students)

#Remove "Charlie" from the dictionary.
students.pop('Charlie',None)
print('After removing Charlie: ',students)

# Print the average score of the remaining students.
avg = sum(students.values())/len(students)
print('Avg num: ',avg)

# Print the names of all students who scored above 85
print('Students who scored above 85: ')
for name,score in students.items():
  if(score>85):
    print(name)
```

```python
# D.2. Write functions to:
inventory ={
  "apple" : (50,0.5),
  "banana" : (100,0.3),
  "orange" : (75,0.7)
}

# Add a new product or update an existing product’s quantity and price.
def add_or_update_product (inventory,name,quantity,price):
  inventory[name] =(quantity,price)
  print(inventory)
  print(f'product:{name}')
  
add_or_update_product(inventory,'mango',30,1.2)
add_or_update_product(inventory,'apple',92,2.3)

# Remove a product by name.
def remove_product(inventory,name):
  if name in inventory:
    inventory.pop(name)
    print(f"product '{name}' removed")
    print("Updated inventory: ",inventory)
  else:
    print(f"product '{name}' not found in inventory")  
    print("Item remains: ",inventory) 
    
remove_product(inventory,'banana') 
remove_product(inventory,'pineapple')   

# Calculate and print the total inventory value (sum of quantity × price for all products).
def total_value(inventory):
  total =0
  for name,(quantity,price) in inventory.items():
    total +=quantity*price
  print("Total value: ",total)  
  
total_value(inventory)  

# List all products whose quantity is below a threshold (e.g., less than 50)
def product_below_threshold(inventory,threshold):
  print(f"products with quantity below {threshold}: ")
  for name,(quantity,price) in inventory.items():
    if(quantity<threshold):
      print(name)
product_below_threshold(inventory,50)
```
